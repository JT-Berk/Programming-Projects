```python
## make a guessing game for a number 1 - 20, let the user determine the number of guesses

# this will be the variable that determines the number of guesses


```

```python
import random

# creates the random varible that user is trying to guess

z = random.randint(1,20) 

# list to append the guesses of the user - just as a reference for the numbers they selected

list1 = []

# while loop to validate user input that the data type is an integer as well as it meets parameters of 1-5 guesses

while (True):
    try:
        x = int(input("How many tries would you like? Enter a number between 1 and 5: "))
    except:
        print("Please choose an integer value")
    else:
        # keeps number in between 1 and 5, zero will not be accepted
        
        if (x <= 5) and (x > 0):
            numberOfTries = x
            # break out of loop if it meets both parameters
            break
        else:
            print("Pick an integer within range")

# validated number moves into next while loop 
# this while loop then prompts the user on the number of guesses they have and begins the guessing process

while (numberOfTries - 1 >= 0):
    print("You have", numberOfTries,"guesses left.")
    while True:
        
        # prompts user to input a number between 1 and 20 - I used a similar structure to that of the number of guesses
        # validation, however there may be a number of better ways to do this
        
        try:
            i = int(input("Guess a number between 1 and 20: "))
        except:
            print("Please choose an integer value")
        else:
            
            # similar structure making sure the users' guess is truly between 1-20 and won't bomb out if 
            # string values are entered
            
            if (i <= 20) and (i >= 0):
                list1.append(i)
                break
            else:
                print("Pick an integer within range")
    
    # if block then tests the number against the randomly generated number
    # starts off by using the correct block so if correct the code doesn't have to iterate unneccesarily
    
    if (i == z):
        print("Correct!")
        print("The correct number was", z,". Your guesses were", list1)
        break
    elif (i > z):
        # number of tries gets reduced each time a number is too high or too low
        
        print("Your number is too high, please try again.")
        numberOfTries -= 1
    elif (i < z):
        print("Your number is too low, please try again.")
        numberOfTries -= 1

# if the number makes it through the loop it was incorrect - the output then gives the user a list of the numbers they guessed
# as well as the correct number
        
if (numberOfTries == 0):
    print("You ran out of guesses :(")
    print("The correct number was", z,". Your guesses were", list1)
```
